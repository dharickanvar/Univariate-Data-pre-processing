```python
#Since getting error in below due to quantitative data is present in dataset getting error to avoid that calling qual,quan fun
```

```python
def quanqual(dataset):
    quan=[]
    qual=[]
    for columnName in dataset.columns:
        #print(columnName)
        if(dataset[columnName].dtype=='O'):
            #print("qual")
            qual.append(columnName)
        else:
            #print("quan")
            quan.append(columnName)
    return quan,qual
```

```python
quan,qual=quanqual(dataset)
dataset[quan]
```
